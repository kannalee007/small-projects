```python
class Factory_type():
    def __init__(self):
        print("-----welcome to factory bonuses--------\n please select the below work type")
        print("1-> permanent")
        print("2-> contract")
        print("3-> intern")
        
        try:
            role = input("please provide your role type:").lower()
            
            if role == 'permanent':
                self.permanent_employee_logic()
            elif role == 'contract':
                self.contractor_logic()
            elif role == 'intern':
                self.Intern_logic()
        
        except ValueError:
            print("Invalid input. Please enter a valid number.")

    def permanent_employee_logic(self):
        try:
            exp = int(input("Ah, you're a permanent employee! What is your number of experience?:"))
            
            base_salary = 0
            if exp < 5:
                base_salary = 50000
            elif 5 <= exp <= 9:
                base_salary = 70000
            elif exp >= 10:
                base_salary = 100000
            
            print(f"Your base salary will be {base_salary}")
            
            if base_salary > 0:
                per = int(input("What is your performance score?:"))
                
                new_salary = base_salary
                
                if per >= 90:
                    new_salary = base_salary * 1.20 
                    print(f"Congratulations! Your new salary with bonus is: {new_salary}")
                elif per > 75 and per <= 89:
                    new_salary = base_salary * 1.05
                    print(f"Your salary with bonus is: {new_salary}")
                else:
                    print("No bonus for this performance score.")
                
                project = 5000
                try:
                    proj = int(input("Number of projects you have done?:"))
                    if proj > 5:
                        print("Your total salary including project bonus will be:", new_salary + project)
                    else:
                        print("Number of projects is too low for a project bonus.")
                except ValueError:
                    print("Invalid number for projects.")
                
        except ValueError:
            print("Invalid input. Please enter a valid number for experience or performance score.")

    def contractor_logic(self):
        try:
            exp = int(input("Ah, you're a contractor! What is your experience?:"))
            
            salary = 0
            if exp >= 5:
                salary = 60000
            elif exp < 5:
                salary = 40000
            
            print(f"Your base salary will be {salary}")
            
            if salary > 0:
                per = int(input("What is your performance score?:"))
                
                new_salary = salary
                
                if per >= 90:
                    new_salary = salary * 1.20 
                    print(f"Yay! Your salary with bonus is {new_salary}")
                elif per>75 and per <= 89:
                    new_salary = salary * 1.10
                    print(f"Yay! Your salary with bonus is {new_salary}")
                else:
                    print("Sorry no bonus")

                project = 3000
                try:
                    proj = int(input("Number of projects you have done?:"))
                    if proj > 3:
                        print("Your total salary including project bonus will be:", new_salary + project)
                    else:
                        print("Number of projects is too low for a project bonus.")
                except ValueError:
                    print("Invalid number for projects.")
                    
        except ValueError:
            print("Invalid input. Please enter a valid number for experience or performance score.")
    def Intern_logic(self):
        try:
            exp = int(input("Ah, you're a intern! What is your performance score?:"))
            stipend = 20000
            if exp >= 90:
                new_sti = stipend * 1.20
                print(f"your new stipend bonus is: {new_sti}")
            elif exp >75 and exp <=89:
                new_sti = stipend * 1.10
                print(f"your new stipend bonus is: {new_sti}")
            else:
                print("sorry no bonus")
        except ValueError:
            print("Invalid input. Please enter a valid number for experience or performance score.")
                      
        
if __name__ == "__main__":
    employee = Factory_type()
```
